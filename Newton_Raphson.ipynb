```python
#Fungsi yang akan digunakan dalam metode Newton-Raphson
def f(x):
    return x**3 - x - 2 # Contoh: f(x) = x^3 - x  - 2

# Turunan dari fungsi f(x)
def f_prime(x):
    return 3*x**2 - 1 # Turunan dari f(x) = 3x^2 - 1
# Metode Newton-Raphson
def newton_raphson (x0, tolerance=1e-6, max_iter=100):
    x = x0 # Titik awal
    iteration = 0

    # Proses iterasi Newton-Raphson
    while iteration < max_iter:
        fx = f(x)
        fpx = f_prime(x)

        # Cek pembagi (derivatif) agar tidak menjadi
        if fpx == 0:
            print("Turunan fungsi f'(x) sama dengan 0. Metode gagal.")
            return None

        # Newton-Raphson formula
        x_new = x - fx / fpx

        # Menampilkan hasil iterasi
        print (f"Iterasi {iteration+1}: x = {x_new}, f(x) = {fx}")

        # Cek konvergensi
        if abs(x_new - x) < tolerance:
            print (f"Konvergen setelah {iteration+1} iterasi.")
            return x_new

        # Update x untuk iterasi berikutnya
        x = x_new
        iteration += 1

    print (f"Metode tidak konvergen setelah {max_iter} iterasi.")
    return None

# Contoh pemanggilan fungsi dengan titik awal x0 = 1
x0 = 1 # Titik awal
root = newton_raphson(x0)

if root is not None:
    print (f"Akarnya persamaan non-linear adalah: {root}")
```

2. Buat grafik pencarian akarnya (silahkan mencari referensi) (40 Point)

```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi Newton-Raphson
def newton_raphson_plot(f, f_prime, x0, tol=0.01, max_iter=10):
    iter_history = []
    x_current = x0

    for _ in range(max_iter):
        f_val = f(x_current)
        f_prime_val = f_prime(x_current)

        if f_prime_val == 0:
            print("Turunan nol. Metode gagal.")
            return None, iter_history

        iter_history.append((x_current, f_val))

        if abs(f_val) < tol:
            break

        x_next = x_current - f_val / f_prime_val
        x_current = x_next

    return x_current, iter_history

# Fungsi untuk Membuat Plot
def plot_newton_raphson(f, f_prime, x0, iter_history, func_label='f(x)'):
    # Rentang untuk plot
    x_vals = np.linspace(x0 - 5, x0 + 5, 400)
    y_vals = f(x_vals)

    plt.figure(figsize=(10, 6))
    plt.plot(x_vals, y_vals, label=func_label, color='blue')
    plt.axhline(0, color='black', linewidth=0.7)

    # Iterasi Newton-Raphson
    for x, y in iter_history:
        plt.plot(x, y, 'ro')  # Titik iterasi
        tangent_x = np.linspace(x - 1, x + 1, 10)
        tangent_y = f(x) + f_prime(x) * (tangent_x - x)
        plt.plot(tangent_x, tangent_y, 'r--', linewidth=0.7)

    plt.title(f'Newton-Raphson Iteration for {func_label}')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid()
    plt.show()

# Fungsi Pertama
def f1(x):
    return x**2 - 2*x - 2

def f1_prime(x):
    return 2*x - 2

# Fungsi Kedua
def f2(x):
    return x**3 - 2*x**2 + 2*x - 5

def f2_prime(x):
    return 3*x**2 - 4*x + 2

# Implementasi untuk Fungsi Pertama
print("Mencari akar untuk f(x) = x^2 - 2x - 2")
akar_f1, iter_f1 = newton_raphson_plot(f1, f1_prime, x0=3)
plot_newton_raphson(f1, f1_prime, x0=3, iter_history=iter_f1, func_label='f(x) = x^2 - 2x - 2')

# Implementasi untuk Fungsi Kedua
print("\nMencari akar untuk f(x) = x^3 - 2x^2 + 2x - 5")
akar_f2, iter_f2 = newton_raphson_plot(f2, f2_prime, x0=2)
plot_newton_raphson(f2, f2_prime, x0=2, iter_history=iter_f2, func_label='f(x) = x^3 - 2x^2 + 2x - 5')
```
